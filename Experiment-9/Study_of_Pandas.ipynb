```python
import pandas as pd

# Create DataFrame
data = {
    "Name": ["Tira", "Vaidehi", "Meera"],
    "Marks": [85, 91, 77]
}

df = pd.DataFrame(data)

# Calculate mode
modedf = df["Marks"].mode()

print("Mode:", modedf)
```

```python
# Aggregate Functions in Data Frames

min_value = df["Marks"].min()
max_value = df["Marks"].max()
sum_value = df["Marks"].sum()

print("Maximum of Marks: ", max_value)
print("Minimum Value of Marks: ", min_value)
print("Sum of Marks: ", sum_value)
```

```python
# Apply conditions (Filtering)
# Display Students who scored more than 90

df[df["Marks"] > 90]
```
