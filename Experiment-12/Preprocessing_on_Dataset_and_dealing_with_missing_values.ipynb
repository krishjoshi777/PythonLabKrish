```python
# Preprocessing on Dataset-1
import pandas as pd

data = {
    "Roll_No":[101,102,103,104],
    "Name":['Harshit',"Kshitij","Khush","Dahiwade"],
    "Age":[18,"-",21,"67"],
    "Marks":[85,90,"-",99],
    "Department":["CSE","IT","E&TC","cse"],
    "Admission_Date":["01-06-2023","03-06-2023","05-06-2023","07-06-2023"]
}
df = pd.DataFrame(data)
#Save as CSV
df.to_csv("student_data.csv",index=False)
df
```

```python
import pandas as pd
import numpy as np
#Step 1
df = pd.read_csv("student_data.csv")
print("Original Data")
print(df)
```

```python
df.replace("-",np.nan, inplace=True)
df
```

```python
# Step 2 : Replace '-' with NaN to mark missing values
df.replace("-",np.nan,inplace=True)
#Convert columns to numeric where needed
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df["Marks"] = pd.to_numeric(df["Marks"], errors='coerce')
'''When converting data (like strings → datetime),
if some values are invalid or cannot be converted,
instead of throwing an error, it will :
Replace those invalid values with NaT (not a time)'''
df
```

```python
# Step 3 : Check missing values
print("\nMissing Values")
print(df.isnull().sum())
```

```python
# fill missing age with mean
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
# Fill missing marks with median
df["Marks"].fillna(df["Marks"].median(),inplace=True)
```

```python
df
```

```python
print(df.isnull().sum())
```

```python
# Step 5 : Fix inconsistent text formatting
df["Department"] = df["Department"].str.upper()
```

```python
# Step 6 : Convert Admission_Date to standard datetime format
df["Admission_Date"]= pd.to_datetime(df["Admission_Date"], errors='coerce')

'''If a value cannot be converted to a valid date, pandas will replace it with NaT (Not a Time) instead of giving an error
Example : invalid dates like "abc", or "32-13-2024" become NaT'''
```

```python
#Step 7 : Show cleaned data
print("\nCleaned Data")
print(df)
```

```python
#Step 8 : Save cleaned dataset
df.to_csv("cleaned_student_data.csv",index=False)
print("\nCleaned CSV file saved as cleaned_student_data.csv")
```

```python
# Working on missing values on Dataset-2
df = pd.read_csv('/content/Cars93.csv')
df
```

```python
df.isna().sum() #Attributes are displayed
```

```python
#Imputing the missing value
#1. Luggage Room
df['Luggage.room']=df['Luggage.room'].fillna(df['Luggage.room'].mean())
```

```python
#2. Rear.seat.room
df['Rear.seat.room']=df['Rear.seat.room'].fillna(df['Rear.seat.room'].mean())
```

```python
#3.AirBags
df['AirBags']=df['AirBags'].fillna(df['AirBags'].mode().values[0])
```

```python
#Save cleaned dataset
df.to_csv("cleaned_Cars93.csv",index=False)
print("\nCleaned CSV file saved as cleaned_Cars93.csv")
```
