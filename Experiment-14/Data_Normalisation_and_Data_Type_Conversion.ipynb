```python
#Data set
import pandas as pd
import numpy as np
data = {
    'Product':['Laptop','Tablet','Headphones','Smartwatch','Camera','Mobile'],
    'Price':[55000,20000,10000,2000,15000,30000],
    'Units_Sold':[25,60,35,80,50,15],
    'Discount':[10,15,12,20,18,8]
}
df = pd.DataFrame(data)
df
```

```python
#Min-Max Normalisation
#Min-Max Normalisation for price

df['Price_MinMax']=(df['Price']-df['Price'].min())/(df['Price'].max()-df['Price'].min())
df
```

```python
# Z-score normalisation
df['Units_Zscore']=(df['Units_Sold']-df['Units_Sold'].mean())/df['Units_Sold'].std()
df
```

```python
#Decimal Scaling
df['Price_Decimal']=df['Price']/10000
df
```

# PART B- Turning categorical variable to quantitative variables in Python


```python
import pandas as pd
import numpy as np
data = {
    'Order_ID':[101,102,103,104,105,106],
    'Customer_Gender':['Male','Female','Male','Female','Male','Female'],
    'Payment_Method':['COD','UPI','Debit Card','UPI','Credit Card','Debit Card'],
    'Product_Category':['Electronics','Clothing','Electronics','Clothing','Electronics','Clothing'],
    'City':['Pune','Delhi','Mumbai','Chennai','Surat','Patiala'],
    'Order_Value':[1500,800,1200,900,2000,1000]
}
df = pd.DataFrame(data)
df
```

```python
#Method 1: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender_Label'] = le.fit_transform(df['Customer_Gender'])
df
```

```python
#Method-2 One-Hot Encoding
df_encoded= pd.get_dummies(df, columns=['Payment_Method'])
df_encoded
```
