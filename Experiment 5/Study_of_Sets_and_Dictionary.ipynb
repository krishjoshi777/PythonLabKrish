# Set


```python
# Creation of Set
myset = {"lion","tiger","monkey","banana"}
print(myset)
```

```python
# Duplicates not allowed
myset = {"lion","tiger","monkey","banana","lion"}
myset
```

```python
# The values of True and 1 are considered the same value in sets, and are treatred as duplicates :
myset = {"lion","tiger","monkey","banana",True,1,2}
print(myset)
myset = {"lion","tiger","monkey","banana",False,0}
print(myset)
```

```python
# A set allows string, integers, boolean values, float etc
newset = {"a",40,1.1,True}
newset
```

```python
# To check membership
myset = {"lion","tiger","monkey","banana"}
print("lion" in myset)
print("monkey" not in myset)
```

```python
# Once a set is created you cannot replace its items, but u can add new items, and remove items
myset = {"lion","tiger","monkey","banana"}
myset.add("elephant")
myset.remove("banana")
myset
```

```python
# Set operations
a={10,20,30,40,50}
b={40,50,60,70,80}
print("Union :",a|b)
print("Intersection :",a&b)
print("Difference :",a-b)
print("Symmetric Difference :",a^b)
```

```python
# Dictionary takes the latest valur given for key. It will not allow duplicate values
mydictionary = {"brand":"Porche","model":"Cayenne","year":2024,"year":2025}
mydictionary
```

```python
# Using len function to find out lenght of a dictionary
mydictionary = {"brand":"Porche","model":"Cayenne","year":2024}
print("Number of members in dictionary :",len(mydictionary))
```
