```python
# Slicing
numbers = [10,20,30,40,50]
print("Elements from index 1 to 3 :",numbers[1:4])
print("First 3 elements :",numbers[:3])
print("Elements from index 2 onwwards : (Reverse indexing)",numbers[-3:])
print("All elements :",numbers[:])
```

```python
#1.
marks = [78,85,90,66,88]
print("Max : ",max(marks))
print("Min : ",min(marks))
print("Sum of marks :",sum(marks))
print("No. of students : ",len(marks))
marks.sort()
print("Ascending order : ",marks)
print("Average marks :",sum(marks)/len(marks))
```

```python
'''2. Grocery Shopping list
A user maintains a grocery list for monthly shopping'''
groc=["Eggs", "Brocolli", "Butter", "Olive", "Oil"]
print("Created gorcery list :",groc)
groc.append("Sugar")
print("Added new items :",groc )
groc.remove("Oil")
print("Remove purchased items : Oil")
print("Final list :",groc)
```

```python
'''3.Contact list
A user maintains a contact list'''
contact=["Krish","Krishiv","Kshitij","Khush","Kahan","Vishvam"]
print("Contacts list created :",contact)
contact.append("Harshit")
print("Add Contact : Harshit")
contact.remove("Krishiv")
print("Remove contact Krishiv :",contact)
contact.sort()
print("Alphabetical order :",contact)
```

```python
#Slicing Practice
list1=[20,30,50,60,40,10,70]
print(list1[1:3])
print(list1[1:3:1])
print(list1[1:5:2])
print(list1[1:4:3])
print(list1[6:4:-2])
print(list1[4:6:-1])
print(list1[-2:-4:-2])
```

```python
'''4. Temperature Analysis
Daily temperature reading of a city are stored'''
temperature=[7,11,20,19,25,14,9,10,21,5]
print("Temprature of first five days :",temperature[0:5],"\nTemprature of last 5 days :",temperature[-5:])
print("Highest temperature :",max(temperature))
print("Lowest temperature :",min(temperature))
print("Average of the temperature :",sum(temperature)/len(temperature))
```

```python
'''5. Daily expense tracker
A person maintains track of daily expense'''
expense = [100,250,150,275,300,500,1000]
print("Total expense :",sum(expense))
print("Average expense :",sum(expense)/len(expense))
print("Maximum expense :",max(expense))
print("Minimum expense :",min(expense))
```
